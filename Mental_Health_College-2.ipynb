**Background** 

Mental health has become a prominent topic of discussion in the United States, as people all over the country have undergone crises with their mental well-being. While mental health disorders vary by each individual, it is important to identify specific groups at risk of experiencing such disorders. A focal point for this research is examining college students who face risks of developing mental health disorders based on their lifestyle, as well as academic and demographic factors. It was found by the American Psychological Association that during the 2020-2021 school year, more than 60 percent of college students met the criteria for at least one mental health problem (Pedrelli et al., 2015). Main areas of focus for mental health with college students can be seen from disorders such as depression, generalized anxiety disorder (GAD), post-traumatic stress disorder (PTSD), obsessive compulsive disorder (OCD), and bipolar disorder (BAD). More specifically, the 2022 Health Minds Survey found that 44% of students reported symptoms of depression, 37% experienced anxiety, and 15% seriously considered suicide, all of which were the survey’s highest rates in results over the last 15 years (Abrams, 2022). When taking account for such metrics, it is crucial to examine the potential influences that higher education can have on students’ relationships with their well-being, as poor mental health has emerged as the leading impediment to academic success (Bowe, 2023).

```python
#import packages 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

```python
# Read the CSV files into a DataFrame
LONELINESS = 'LonelinessScale.csv'
PHQ = 'PHQ-9.csv'
loneliness = pd.read_csv(LONELINESS)
phq = pd.read_csv(PHQ)
GRADES = 'grades.csv'
grades = pd.read_csv(GRADES)
```

```python
# Define values dictionary
values = {"Sometimes": 0.7, "Often": 1, "Rarely": 0.3, "Never": 0, "Not at all": 1, "Several days": 0, "Nearly every day": 0.2, "More than half the days": 0.5}
```

**Functions**
There are several functions that are needed in order to evaluate this data. 
1. The first function will read the csv files and give the different files names that can be called upon.
2. The second function will map values associated to specific phrases for responses within the files.
3. The third function split the loneliness and depression data into two parts: pre-experiment and post-experiment.
4. The fourth function will plot the various items called up to make a scatter plot with a linear line connecting the dots for loneliness score and depression score.
5. The fifth function will graph a regression plot with a correlation line of best fit for the items called upon such as loneliness score, depression score, and grades.

```python
df_phq_2 = phq.copy()
for col in df_phq_2.columns[2:]:
    df_phq_2[col] = df_phq_2[col].map(values)
```

```python
# Drop the rows that have only the pre data  
df_student_lon_pre = df_student_lon.drop(df_student_lon[df_student_lon['type'] == 'post'].index)
df_phq_pre = df_phq.drop(df_phq[df_phq['type'] == 'post'].index)
```

```python
#Drop the rows that have only the post data
df_student_lon_post = df_student_lon_2.drop(df_student_lon_2[df_student_lon_2['type'] == 'pre'].index)
df_phq_post = df_phq_2.drop(df_phq_2[df_phq_2['type'] == 'pre'].index)
```

```python
# Compute the wellness score as a nuneric value for pre
row_sums_lon_pre = df_student_lon_pre.iloc[:, 2:].sum(axis=1)
df_student_lon_pre['loneliness_score'] = row_sums_lon_pre
row_sums_phq_pre = df_phq_pre.iloc[:, 2:].sum(axis=1)
df_phq_pre['depression_score'] = row_sums_phq_pre
```

```python
# Compute the wellness score as a nuneric value for post
row_sums_lon_post = df_student_lon_post.iloc[:, 2:].sum(axis=1)
df_student_lon_post['loneliness_score'] = row_sums_lon_post
row_sums_phq_post = df_phq_post.iloc[:, 2:].sum(axis=1)
df_phq_post['depression_score'] = row_sums_phq_post
```

```python
# Restructure the uid such that it is only numeric for pre
df_student_lon_pre['uid'] = df_student_lon_pre['uid'].astype(str)
df_student_lon_pre['uid'] = df_student_lon_pre['uid'].str.replace('u', '')

df_phq_pre['uid'] = df_phq_pre['uid'].astype(str)
df_phq_pre['uid'] = df_phq_pre['uid'].str.replace('u', '')
```

```python
# Restructure the uid such that it is only numeric for post
df_student_lon_post['uid'] = df_student_lon_post['uid'].astype(str)
df_student_lon_post['uid'] = df_student_lon_post['uid'].str.replace('u', '')

df_phq_post['uid'] = df_phq_post['uid'].astype(str)
df_phq_post['uid'] = df_phq_post['uid'].str.replace('u', '')
```

```python
#Plot loneliness score for type pre
plt.figure(figsize=(18, 7))
plt.plot(df_student_lon_pre['uid'].values, df_student_lon_pre['loneliness_score'].values, marker='o', linestyle='-', color="blue")
plt.title('Graph 1: Pre-Experiment Loneliness Score vs UID')
plt.xlabel('UID')
plt.ylabel('Loneliness Score')
plt.grid(True)
plt.show()
```

```python
#Plot loneliness score for type only as post
plt.figure(figsize=(18, 7))
plt.plot(df_student_lon_post['uid'].values, df_student_lon_post['loneliness_score'].values, marker='o', linestyle='-', color="blue")

plt.title('Graph 2: Post-Experiment Loneliness Score vs UID')
plt.xlabel('UID')
plt.ylabel('Loneliness Score')
plt.grid(True)
plt.show()
```

**Discussion and Limitations**

This study helps to unveil crucial insights regarding the correlations between loneliness, depression, and academic performance, as well as the inherent limitations of such. A primary constraint lies in the study's reliance on a specific sample size, potentially undermining the broader scope of the findings. The absent component of diversity within this cohort raised questions about the general applicability of the results, emphasizing the need for caution in extrapolating these findings to a more diverse student population. Another limitation presents itself in the form of self-reporting bias. This can be seen as a common obstacle in research focused on mental health. The reliance on participants' subjective accounts imposes a risk for potential response bias, skewing the data and compromising the study's overall validity and reputability. Temporal considerations emerge as an additional limitation, with the study capturing data at only two time points. A more nuanced examination of temporal fluctuations could enrich our understanding of how loneliness and depression evolve over specific periods, adding depth to the study's temporal research. 

Sociodemographic factors, including age, gender, and socioeconomic status, constitute an additional limitation, as the study does not explicitly consider their influence on the observed correlations. These factors, paramount in shaping individual experiences, demand attention for a more comprehensive understanding. Further limitations pertain to the singular metric employed to represent academic performance – 'gpa all.' This oversimplification neglects the diverse facets of academic success, failing to account for variations across courses, majors, or learning styles that may impact the relationship between depression and academic performance in nuanced ways. It is also important to note study's adherence to correlations without establishing causation. While correlations are valuable, discerning causative relationships necessitates a more profound exploration of potential confounding variables or external factors that may influence observed outcomes.

Arguably the most relevant limitation within upon conducting this study was the inability to utilize sleep data from the StudentLife study, as the dataset presented itself as insufficient. Had this dataset been deemed more reliable, this could have provided a potentially relevant component to the study’s research into additional components at play in students’ mental well-being. In acknowledging these limitations, the study lays the groundwork for future research endeavors, underscoring the necessity for a more diverse and representative sample, transparent reporting of experimental details, and a nuanced exploration of temporal, sociodemographic, and contextual factors. These considerations will not only refine our understanding of students’ well-being but also guide the development of more targeted interventions and support systems within academic settings.
